# **Задание №6. Разработка спецификации требований к информационной системе (SRS)**

#### **1. ВВЕДЕНИЕ**

##### **1.1. Цель документа**

Данный документ определяет функциональные и нефункциональные требования к десктопному приложению «Аэродинамический симулятор автомобиля», разрабатываемому в рамках дипломного проекта для использования в учебном процессе НАМИ. Цель приложения — наглядная демонстрация влияния аэродинамических модификаций (спойлеры, диффузоры и др.) на аэродинамические и скоростные характеристики автомобиля.

##### **1.2. Область применения**

Приложение предназначено для:

- студентов-практикантов НАМИ — для изучения основ автомобильной аэродинамики и экспериментирования с модификациями;
- преподавателей НАМИ — для демонстрации на занятиях и проверки результатов студентов;
- широкой аудитории (автолюбители, школьники) — для самообразования в области инженерной физики.

Приложение не заменяет CFD-расчёт, а служит учебным инструментом для понимания связей между формой, аэродинамикой и динамикой автомобиля.

#### **2. ФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ**

##### **2.1. Управление базой автомобилей и элементов**

**ФТ-001: Выбор базовой модели автомобиля**
- **Описание:** Пользователь может выбрать автомобиль из списка доступных моделей.
- **Входные данные:** Список автомобилей (DrivAer Fastback, Notchback, Estate и др.)
- **Выходные данные:** Отображение 3D-модели, базовых характеристик (Cx, мощность, масса и др.)
- **Бизнес-правила:** Каждый автомобиль имеет уникальный набор параметров.
- **Приоритет:** Критический
- **Критерий проверки:** Выбранный автомобиль отображается в 3D-окне, его параметры корректно загружаются.

**ФТ-002: Управление аэродинамическими элементами**
- **Описание:** Пользователь может добавлять/удалять элементы (спойлер, диффузор и др.) и настраивать их параметры (угол, размер).
- **Входные данные:** Тип элемента, угол (градусы), масштаб (0.5–1.5)
- **Выходные данные:** Обновлённая конфигурация, пересчитанные характеристики
- **Бизнес-правила:** Угол и размер должны быть в допустимых пределах.
- **Приоритет:** Критический
- **Критерий проверки:** Элемент визуально отображается на модели, его параметры влияют на расчёт.

##### **2.2. Расчёт и отображение характеристик**

**ФТ-003: Расчёт аэродинамических характеристик**
- **Описание:** Система мгновенно оценивает Cx и CL с учётом установленных модификаций с помощью нейросетевой модели.
- **Входные данные:** Параметры автомобиля + список модификаций
- **Выходные данные:** Cx, CL, MAE модели ≤ 0.005
- **Приоритет:** Критический
- **Критерий проверки:** Результат соответствует данным DrivAerML в пределах погрешности.

**ФТ-004: Расчёт скоростных характеристик**
- **Описание:** Система рассчитывает максимальную скорость и время разгона 0–100 км/ч на основе Cx, мощности и массы.
- **Выходные данные:** max_speed_kmh, acceleration_0_100
- **Приоритет:** Высокий
- **Критерий проверки:** Расчёт соответствует физической формуле:

##### **2.3. Визуализация и отчётность**

**ФТ-005: 3D-визуализация ламинарных потоков**
- **Описание:** Отображение упрощённых линий обтекания вокруг автомобиля.
- **Выходные данные:** Анимированное 3D-окно с линиями потока
- **Приоритет:** Критический
- **Критерий проверки:** Поток визуально огибает модель, изменяется при добавлении элементов.

**ФТ-006: Сравнение конфигураций**
- **Описание:** Пользователь может сравнить «до» и «после» установки элементов.
- **Выходные данные:** Таблица с разницей по Cx, скорости и т.д.
- **Приоритет:** Высокий
- **Критерий проверки:** Разница отображается корректно, подсвечивается улучшение/ухудшение.

**ФТ-007: Экспорт отчёта в PDF**
- **Описание:** Генерация PDF-отчёта с описанием модификаций и результатов.
- **Выходные данные:** Файл report_YYYYMMDD.pdf
- **Приоритет:** Средний
- **Критерий проверки:** Файл открывается, содержит все данные и скриншот 3D-модели.
