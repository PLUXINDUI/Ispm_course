## 1. СТРАТЕГИЯ ТЕСТИРОВАНИЯ

### 1.1. Цели и задачи
- Обеспечить соответствие приложения функциональным и нефункциональным требованиям из SRS (Задание №6).
- Подтвердить корректность работы нейросетевой модели на валидационной выборке DrivAerML.
- Проверить интерактивность и стабильность 3D-визуализатора на типовых ПК.
- Убедиться, что сценарии студентов-практикантов НАМИ выполняются без ошибок.

### 1.2. Область тестирования

**Включено:**
- Функциональное тестирование: выбор авто, установка элементов, расчёт $C_x$ / $C_L$, визуализация, экспорт отчёта.
- Тестирование нейросетевой модели: точность предсказаний, время отклика.
- Тестирование 3D-визуализатора: корректность отображения линий потока.
- Тестирование UI: простота, интуитивность, производительность.
- Регрессионное тестирование после доработок.

**Исключено:**
- Аудит качества кода нейросети (используется предобученная модель).
- Тестирование на macOS/Linux (только Windows 10/11).
- Нагрузочное тестирование (приложение однопользовательское).

### 1.3. Виды тестирования
- Функциональное (ручное) — основные пользовательские сценарии.
- Модульное — для компонентов: AeroPredictor, VisualizationEngine.
- Производительности — время расчёта, загрузки приложения.
- Юзабилити — прохождение сценариев студентами без инструкции.
- Регрессионное — после каждой доработки.

### 1.4. Инструменты тестирования

| Тип тестирования | Инструмент | Обоснование |
|------------------|------------|-------------|
| Управление тест-кейсами | Excel / TestRail | Простота, интеграция с отчётом |
| Модульное тестирование | pytest | Лёгкость, поддержка Python |
| Тестирование производительности | Встроенные таймеры Python | Достаточно для desktop-приложения |
| Юзабилити-тестирование | Студенты-практиканты НАМИ | Целевая аудитория = тестировщики |
| Визуальная проверка 3D | Ручная проверка на эталонных моделях | Open-source альтернатив нет |

### 1.5. Критерии входа и выхода

**Entry Criteria (условия начала тестирования):**
- Все модули приложения интегрированы.
- Собрана установочная версия .exe.
- Подготовлены тестовые данные (автомобили, элементы, сценарии).

**Exit Criteria (условия завершения):**
- 100% покрытие критических требований тестами.
- Pass Rate ≥ 95% для всех функциональных тестов.
- MAE модели на тестовой выборке ≤ 0.005.
- Время расчёта ≤ 0.5 сек.
- Все критические дефекты исправлены.
- Получена положительная обратная связь от ≥5 студентов.
